```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
import numpy as np

marcadores =np.where(vendas == np.min(vendas),'o','')

plt.plot(meses,vendas, linestyle='--',color='green', linewidth=1)

for posicao, marcador in enumerate(marcadores):
    if marcador =='o':
        cor ='red'
    else:
        cor='blue'
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,color = cor)

plt.annotate("Menor Valor",color ='red',xy=(2, 78),xytext=(3, 80),arrowprops=dict(facecolor='red'))
plt.title("Crescimento das Vendas ao Longo do Ano")
plt.grid()
plt.xlabel("MÃªses")
plt.ylabel("Vendas(Em milhares de R$)")
plt.show()
```
