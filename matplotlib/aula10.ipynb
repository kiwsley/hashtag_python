```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
import numpy as np
from cycler import cycler

marcadores =np.where(vendas == np.min(vendas),'o','')
#paleta de cores
cores = plt.get_cmap('Dark2').colors
ciclo_cores = cycler('color',cores)
plt.rc('axes',prop_cycle = ciclo_cores)
#grafico
plt.plot(meses,vendas, linestyle='--', linewidth=1.5)
#marcadores
for posicao, marcador in enumerate(marcadores):
    if marcador =='o':
        cor ='red'
    else:
        cor='black'
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,color = cor)
    plt.annotate(vendas[posicao],color =cor,xy=(meses[posicao],vendas[posicao]),xytext=(-5,5),textcoords= 'offset points')
    
#plt.annotate("Menor Valor",color ='red',xy=(2, 78),xytext=(3, 80),arrowprops=dict(facecolor='red'))


#destaque de região
plt.axvspan(6,11, alpha = 0.3, color=cores[1])
plt.text(6,80,'Recuperação de Crescimento',color='green', fontsize=9 ,weight='bold')
plt.box(False)
plt.tick_params(axis='x', length =0)
plt.yticks([])
plt.title("Crescimento das Vendas ao Longo do Ano")
plt.grid()
plt.xlabel("Mêses")
plt.ylabel("Vendas(Em milhares de R$)")
plt.show()
```
