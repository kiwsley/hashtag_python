```python
import matplotlib.pyplot as plt
from matplotlib import colormaps 
from cycler import cycler

ano=[2018,2019,2020,2021]
vendas_online = [5,10,15,20]#milhoes de reais
vendas_lojas =[10,8,6,4] #milhoes de reais

cores_cycler= cycler('color', plt.get_cmap('Pastel2').colors)#passando a paleta de cores para matplolib
plt.rc('axes', prop_cycle=cores_cycler)

plt.bar(ano,vendas_online,label='Online') #Utilizando a tupla
plt.bar(ano, vendas_lojas,bottom=vendas_online,label='Loja')
plt.title('Vendas Online VS Lojas')
plt.xlabel('Ano')
plt.ylabel("Vendas em Milh√µes de R$")
plt.xticks(ticks=ano)
plt.legend()
plt.show()
```
