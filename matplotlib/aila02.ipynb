```python
import matplotlib.pyplot as plt
from matplotlib import colormaps as cmap

categorias =['Eletronicos','Roupas','Alimentos', 'Livros']
quantidades =[45,30,15,10]
destaque=[0,0,0.1,0.1]
cores = plt.get_cmap('Pastel2').colors

plt.pie(quantidades,labels=categorias,autopct='%.1f',explode=destaque,colors=cores)
plt.title("Distribuição Percentual%")
plt.show()
```
