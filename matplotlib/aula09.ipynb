```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# listas com os dados
fabricante = ['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Outros']
participacao = [30.2, 26.6, 10.4, 9.7, 4.8, 18.3]
```

```python
#grafico de pizza
from cycler import cycler
cores = plt.get_cmap('Set3').colors
ciclo_cores = cycler('color',cores)
plt.rc('axes', prop_cycle=ciclo_cores)

destaque=[0.1,0.05,0,0,0,0]
plt.pie(participacao, labels= fabricante,autopct='%.2f',explode=destaque,shadow=True)
plt.title('Participação Percetual de Mercado')
plt.show()
```
