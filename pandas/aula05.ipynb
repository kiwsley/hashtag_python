```python

import pandas as pd
```

```python
vendas_df = pd.read_csv('Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv("Contoso - Cadastro Produtos.csv", sep=';',encoding='ISO-8859-1')
lojas_df = pd.read_csv("Contoso - Lojas.csv", sep=';',encoding='ISO-8859-1')
clientes_df = pd.read_csv("Contoso - Clientes.csv", sep=';',encoding='ISO-8859-1')
```

```python
display(vendas_df)
display(produtos_df)
display(lojas_df)
display(clientes_df)
```

```python
clientes_df = clientes_df.drop(['Unnamed: 7','Unnamed: 8','Unnamed: 9', 'Unnamed: 10'],axis =1)
display(clientes_df)
```

```python
clientes_df=clientes_df[['ÿID Cliente', 'E-mail']]
produtos_df=produtos_df[['ID Produto','ÿNome do Produto']]
lojas_df=lojas_df[['ÿID Loja','Nome da Loja']]
display(clientes_df)
```

```python
lojas_df=lojas_df.rename(columns={'ÿID Loja': 'ID Loja'})
clientes_df=clientes_df.rename(columns={'ÿID Cliente': 'ID Cliente'})
```

```python
# Juntas os datas frames

vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df= vendas_df.merge(lojas_df, on='ID Loja')
vendas_df=vendas_df.merge(clientes_df, on='ID Cliente')
```

```python
display(vendas_df)
```

```python
frequecia_clientes=vendas_df['E-mail'].value_counts()
display(frequecia_clientes)
```

```python
#clientes que mais comprou
frequecia_clientes[:5].plot(figsize=(15,5))
```

```python
vendas_lojas = vendas_df.groupby('Nome da Loja').sum()
vendas_lojas=vendas_lojas[['Quantidade Vendida']]
display(vendas_lojas)
```

```python
#ordenar o data frame

vendas_lojas = vendas_lojas.sort_values('Quantidade Vendida', ascending=False)
display(vendas_lojas)
#vendas plotar

vendas_lojas[:5].plot(figsize=(15,5),kind='bar')
```

```python
#maior valor 

maior_valor = vendas_lojas['Quantidade Vendida'].max()
melhor_loja = vendas_lojas['Quantidade Vendida'].idxmax()
print(maior_valor,melhor_loja)
```

```python
menor_valor = vendas_lojas['Quantidade Vendida'].min()
pior_loja = vendas_lojas['Quantidade Vendida'].idxmin()
print(pior_loja,menor_valor)
```

```python
#porcentagem devolvida
qtde_vendida= vendas_df['Quantidade Vendida'].sum()
qtde_devolvida= vendas_df['Quantidade Devolvida'].sum()
print('{:.2%}'.format(qtde_devolvida/qtde_vendida))
```

```python
vendas_lojacontoso = vendas_df[vendas_df['ID Loja'] == 306]
display(vendas_lojacontoso)
```

```python
#porcentagem devolvida loja contoso europe Online ID 306
qtde_vendida= vendas_lojacontoso['Quantidade Vendida'].sum()
qtde_devolvida= vendas_lojacontoso['Quantidade Devolvida'].sum()
print('{:.2%}'.format(qtde_devolvida/qtde_vendida))
```

```python
vendas_lojacontoso_sdevolucao = vendas_df[(vendas_df['ID Loja']==306)  & (vendas_df['Quantidade Devolvida']==0)]
display(vendas_lojacontoso_sdevolucao)
```

```python
# convertendo a coluna data da venda
vendas_df['Data da Venda'] = pd.to_datetime(vendas_df['Data da Venda'],format='%d/%m/%Y')
#criando colunas
vendas_df['Ano da Venda'] = vendas_df['Data da Venda'].dt.year
vendas_df['Ano da mes'] = vendas_df['Data da Venda'].dt.month
vendas_df['Ano da dia'] = vendas_df['Data da Venda'].dt.day
display(vendas_df)
vendas_df.info()
```
